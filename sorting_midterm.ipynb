Merge Sort is a sorting algorithm that employs the Divide and Conquer method. Merge Sort was invented by Dr. John von Neumann, Ph.D, in 1945. Dr. von Neumann was involved in the Manhattan Project for the duration of the project, inlcuding through 1945. The original physical copy of the code has visible indentations where "Top Secret" was once written and later erased.  It is very possible, and even probable that the development of Merge Sort, much like the development of the Monte Carlo method, is a byproduct of the Manhattan project. 

Merge sort works by repetitively dividing an unsorted list or array until temporary working lists or arrays with length of one are obtained.  Since these working lists only contain one item they are considered sorted.  Then these working lists are sucessively merged back together, evaluating to determine order as they are merged, until a final list containing all of the original elements of the list, in the new order is made.

The implementation below is based off an approach modeled in the Runestone Data Structures Online Textbook, as well as on GeeksForGeeks. In my implementation I have modified this approach by seperately defining a function to merge the working arrays back together. My mergesort function takes an array of numbers, determines the index of the midpoint of the array and breaks the array into two halves, Left and Right. Then my mergesort function is called recursively on the Left half.  This traverses depthwise until two subarrays with lengths of one are obtained.  Then my merge function is called to rejoin the two subarrays, evaluating for size and placing the smaller number first. This repeats until all of the items in the left sublist have been sorted into subarrays with length of one and merged back together.  Then my mergesort function repeats this process for the Right side of the array. Once the right side of the array has been sorted and merged it is merged with the Left side of the array to create a final array containing all of the elements of the original array, in ascending numerical order.  

```python
import random
import time 
import pandas as pd 
import altair as alt
import numpy as np
import math
from multiprocessing import Manager, Pool, Process, cpu_count
import multiprocessing
import os
import functools
import _thread
from contextlib import contextmanager

def get_list(n):
    """ Generate a list of n random numbers """
    #use random sample to make a list of length n of random numbers less than 100 
    # This is my code from Assignment 2
    mylist = random.sample(range(0, 10000), n)
    return mylist
    print(mylist)
    
def merge(anarray, left, right): 
    """This is a function to merge two arrays containing numbers together"""
    l = 0 #left side iterator
    r = 0 #right side iterator
    a = 0 # array iterator 
    
    while l < len(left) and r < len(right):
        if left[l] <= right[r]: #if left is less than or equal to right, add left to array
            anarray[a]=left[l]
            l = l+1 #increment left
        elif left[l]>right[r]: #if left is greater than right add right to array
            anarray[a]=right[r]
            r=r+1 #increment right
        a=a+1 #increment array
        
    while l <len(left): 
        anarray[a]= left[l]
        l = l+1 #increment left
        a = a+1 #increment array
    while r <len(right):
        anarray[a]= right[r] 
        r = r+1 #increment right
        a = a+1 #increment array 
        
def mergesort(anarray):
    """This is a function to sort an array using the Merge Sort method"""
    if len(anarray)>1:
        middle = int((len(anarray))/2)
    
        Left = anarray[middle:] #array of items to the left of midpoint
        Right = anarray[:middle] #array of items to the right of midpoint 
        
        mergesort(Left) #recursively sort both halves
        mergesort(Right)
        merge(anarray, Left, Right)#merge working arrays together
    
            
mylist= get_list(10)    
mergesort(mylist)
print(mylist)
```

```python
def mergesort2(anarray):
    """This is a duplicate of mergesort with a print statement
        to demonstrate the Divide and Conquer Method """
    if len(anarray)>1:
        middle = int((len(anarray))/2)
    
        Left = anarray[middle:] #array of items to the left of midpoint
        Right = anarray[:middle] #array of items to the right of midpoint 
        
        mergesort2(Left) #recursively sort both halves
        mergesort2(Right)
        merge(anarray, Left, Right)
    print(anarray)
mylist2= get_list(10)    
mergesort2(mylist2)
```

The above version of mergesort with a print statement is to demostrate the depth first sorting strategy implemented in mergesort. 

```python
def get_time_elapsed(function_name, threshold=0.05):
    """ Return time taken to execute a function """
    # This is my code from Assignment 2 
    timestorun = 1
    while True:
        start_time = time.time()              
        for i in range(timestorun):
            function_name()
        end_time= time.time()
        time_elapsed = end_time- start_time    
        if  time_elapsed> threshold:
            return time_elapsed/timestorun
        else:
            timestorun = timestorun*2
```

#### Parallel versus Traditional Merge Sort
Traditional Sequential Merge Sort works by first sorting and merging all of the elements in one temporary sublist and then proceeding to address the elements in the other temporary sublist.  Parallel implementations of merge sort circumvent this and perform either both or one of these operations in parallel on seperate processors. I attempted to create a parrallel implementation of Merge sort in python to split and simutaineously sorts the left and right sides of the array. This attempted implementation was based off of the pseudo code in the Wikipedia Article about Merge Sort. Unfortunately multiprocessing in python with recursion is difficult to implement, and I was unable to find a functional method. 

However, I was able to find a multiprocessing API with shared memory in C which is easier to implement. I have modified an implementation of Merge Sort found on GeeksForGeeks to use the Open Multi Processing(Open MP) API. This modified implementation now sorts the left and right sides of the array in parallel and is uploaded to the assignment portal as parallel_merge_sort.c. 

```python
def parallel_mergesort(anarray):
    """Non Functional Implementation of Merge Sort with parallel recursion"""
    #This is one of my attempts to implement a parallel merge sort in python, 
    # I included it because even though it is non-functional, it is close to the psuedo code found on Wikipedia
    # and illustrates the Fork-Join method more clearly than my functional C implementation.  
    if len(anarray)>1:
        middle = int((len(anarray))/2)
    
        Left = anarray[middle:] #array of items to the left of midpoint
        Right = anarray[:middle] #array of items to the right of midpoint 
        

        #recursively sort both halves in parallel 
        left_process = Process(target=parallel_mergesort, args=(Left,))
        right_process = Process(target=parallel_mergesort, args=(Right,))
        #set up seperate process and start
        left_process.start()
        right_process.start()

        left_process.join() #rejoin
        right_process.join()
        
        
        merge(anarray, Left, Right)#merge working arrays together
        
#mylist3= get_list(10)    
#parallel_mergesort(mylist3)
#print(mylist3)
```
